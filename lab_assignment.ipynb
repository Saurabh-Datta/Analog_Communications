```
import numpy as np
import matplotlib.pyplot as plt
import math
```

**Exercise 1**

**Sinusoidal signal**

```
pi = math.pi
t = np.arange(0,4*pi,pi/1000)
x_ticks=[0, pi/2,pi,3*pi/2,2*pi,5*pi/2,3*pi,7*pi/2,4*pi]
x_labels=['0','π/2','π','3π/2','2π','5π/2','3π','7π/2','4π']
a = np.sin(t)
plt.plot(t,a)
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.show()
```

**Cosine signal with three different amplitudes on same graph**

```
b1 = np.cos(t)
b2 = 1.5*b1
b3 = 2*b1
plt.plot(t, b1, t, b2, t, b3)
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.show()
```
