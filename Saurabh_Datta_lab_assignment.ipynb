# **Experiment 1**

To perform fourier synthesis of 



1.   Square Wave
2.   Triangular Wave



1. Square Wave

Normal Synthesis

```
#normally synthesizing square wave by using in-built functions
sq = sg.square(pi*t)
plt.plot(pi*t,sq)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.ylim(-2,2)
plt.show()
```

Fourier Synthesis

```
#Adding the terms according to the fourier series derived for square wave
k=1
s = np.zeros((1,len(t)))
while k<100001:
  temp1 = np.sin(k*pi/2)
  temp2 = np.cos((k/2)*(pi-(2*pi*t)))
  s = s + 4*temp1*temp2/(pi*k)
  k = k+1
```

```
#reshaping it for programming etiquettes
np.reshape(s,(4000,))
```

```
#initializing the first element
s[0,0]=s[0,1]
```

```
#plotting the series
plt.plot(pi*t,s[0])
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.ylim(-2,2)
plt.show()
```

```
#plotting the graphs together for comparision
plt.figure(1,figsize=(10,8))
plt.subplot(211)
plt.title("Square Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.ylim(-2,2)
plt.plot(pi*t, sq)
 
plt.subplot(212)
plt.title("Fourier Synthesis")
plt.plot(pi*t, s[0])
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.tight_layout()
plt.ylim(-2,2)
plt.show()
```

2. Triangular Wave

Normal Synthesis

```
#normally synthesizing triangular wave by using in-built functions
tn = sg.sawtooth(pi*t,0.5)
plt.plot(pi*t,tn)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.ylim(-2,2)
plt.show()
```

Fourier Synthesis

```
#adding the series as derived for triangular wave 
k=1
s = np.zeros((1,len(pi*t)))
while k<100001:
  temp1 = (-8/((k**2)*(pi**2)))
  temp2 = np.cos(k*pi*t)
  s = s + (temp1*temp2)
  k = k+2
```

```
#Reshaping the series as per programming etiquettes
np.reshape(s,(4000,))
```

```
#plotting the series
plt.plot(pi*t,s[0])
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.ylim(-2,2)
plt.show()
```

```
#plotting both graphs together for comparision
plt.figure(1,figsize=(10,8))
plt.subplot(211)
plt.title("Triangular Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.ylim(-2,2)
plt.plot(pi*t, tn)
 
plt.subplot(212)
plt.title("Fourier Synthesis")
plt.plot(pi*t, s[0])
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xticks(ticks=x_ticks,labels=x_labels)
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.tight_layout()
plt.ylim(-2,2)
plt.show()
```

#**Experiment 3**

**To study the working of AM reciever**

Detection of AM Wave by Envelope Detector

```
from scipy.signal import hilbert
```

```
#Detection of am wawes by envelope detector
Ac=2     #carrier amplitude
fc=0.5   #carrier frequency
Am=.5    #message signal amplitude
fm=.05   #message signal frequency
Fs=100   #sampling rate/frequency

ka=1     #modulation coefficient

ct=Ac*np.cos(2*pi*fc*t) #defining the carrier signal wave
mt=Am*np.cos(2*pi*fm*t) #defining the message signal
AM=ct*(1+ka*mt)         #Amplitude Modulated wave, according to the standard definition

plt.plot(t, AM)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.axhline(y=0, color='k')
plt.grid(True, which='both')

envelope = abs(hilbert(AM))  #hilbert transform to calculate the envelope of the signal
plt.plot(t, envelope)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.show()
```

```
plt.figure(1,figsize=(10,8))
plt.subplot(311)
plt.plot(t, sg.square(2 * np.pi * 5 * t))
plt.title("Pulse Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.xlim(0,2)
plt.ylim(-2, 2)
plt.show()

sig = np.sin(2 * np.pi * t)
pwm = sg.square(2 * np.pi * 30 * t, duty=(sig + 1)/2)
plt.figure(1,figsize=(10,8))
plt.subplot(312)
plt.plot(t, sig)
plt.title("Message Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.xlim(0,2)
plt.show()

plt.figure(1,figsize=(10,8))
plt.subplot(313)
plt.plot(t, pwm)
plt.title("PWM Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.axhline(y=0, color='k')
plt.grid(True, which='both')
plt.ylim(-1.5, 1.5)
plt.xlim(0,2)
plt.show()
```
